```python
import torch
import torchaudio
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(torch.__version__)
print(torchaudio.__version__)

torch.random.manual_seed(0)
device = "cuda" if torch.cuda.is_available() else "cpu"

out_path = '../../data/ATIS/raw/test/ASR_voice_samples/'
```

```python
bundle = torchaudio.pipelines.TACOTRON2_WAVERNN_PHONE_LJSPEECH

processor = bundle.get_text_processor()
tacotron2 = bundle.get_tacotron2().to(device)
vocoder = bundle.get_vocoder().to(device)
```

```python
data = pd.read_csv('../../data/ATIS/experiments/clean/test/test.tsv',sep='\t')
data = list(data['TEXT'])
```

```python
sent = ['could you please book a flight to las vegas','i am going to home']
with torch.inference_mode():
    processed, lengths = processor(sent)
    processed = processed.to(device)
    lengths = lengths.to(device)
    spec, spec_lengths, _ = tacotron2.infer(processed, lengths)
    waveforms, lengths = vocoder(spec, spec_lengths)
```

```python
torchaudio.save(out_path+"/{}.wav".format(3), waveforms[0:1].cpu(), sample_rate=vocoder.sample_rate)
torchaudio.save(out_path + "/{}.wav".format(4), waveforms[1:2].cpu(), sample_rate=vocoder.sample_rate)
```

```python
for idx,sent in data:
    
    with torch.inference_mode():
        processed, lengths = processor(sent)
        processed = processed.to(device)
        lengths = lengths.to(device)
        spec, spec_lengths, _ = tacotron2.infer(processed, lengths)
        waveforms, lengths = vocoder(spec, spec_lengths)

    torchaudio.save(out_path + "/{}.wav".format(idx), waveforms[0:1].cpu(), sample_rate=vocoder.sample_rate)
```
