```python
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('../data/multiATIS/raw/train_dev_test/train_EN.tsv',sep='\t')
data
```

```python
slots = set(list(data.intent))
slots
```

```python
slot_dict = []
for slot_seq in slots:
    slot_tokens = slot_seq.split(' ')
    for t in slot_tokens:
        slot_dict.append(t)
```

```python
slot_dict
Counter(slot_dict)
```

```python
# sentence length
textData = list(data.TEXT)
lengths = []

for sent in textData:
    lengths.append(len(sent.split(' ')))

hist = Counter(lengths)
```

```python
for key,value in hist.items():
    
    if key>=14:
        print(value)
```
