```python
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('../data/multiATIS/split/train/train_EN.tsv',sep='\t')
data
```

```python
slots = list(data.SLOTS)
slots
```

```python
slot_dict = []
for slot_seq in slots:
    slot_tokens = slot_seq.split(' ')
    for t in slot_tokens:
        slot_dict.append(t)
```

```python
slot_dict
Counter(slot_dict)
```
